```python
df_stock= pd.read_csv('SP500_close_price_no_missing.csv')
df_stock=pd.DataFrame(df_stock)
```

```python
import pandas as pd
import numpy as np

# Assuming df_stock is your original DataFrame

# Create a new DataFrame to store log returns
log_returns_df = pd.DataFrame()

# Include the date column in the new DataFrame
log_returns_df['date'] = df_stock['date']

# Iterate through columns (excluding the first, assuming it's the date column)
for column in df_stock.columns[1:]:
    # Calculate log returns for the current column
    log_returns = np.log(df_stock[column].shift(-1) / df_stock[column])
    
    # Store log returns in the new DataFrame
    log_returns_df[f'{column}_Log_Return'] = log_returns

# Drop the last row since the last log return will be NaN
log_returns_df = log_returns_df.dropna()

# Display the DataFrame with log returns
log_returns_df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame

# Extract numerical columns (excluding non-numeric columns)
numeric_columns = log_returns_df.select_dtypes(include=[np.number]).columns

# Create a StandardScaler object
scaler = StandardScaler()

# Standardize each value in the numeric columns
log_returns_df[numeric_columns] = scaler.fit_transform(log_returns_df[numeric_columns])

# Display the standardized DataFrame
log_returns_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and contains standardized prices
# Replace 'your_dataset.csv' with the actual file or use your DataFrame
df= log_returns_df

# Select columns for which you want to compute the covariance matrix
# Here, I'm assuming you want to compute the covariance matrix for standardized prices
covariance_columns = df.columns[df.columns.str.endswith('_Return')]

# Extract the relevant subset of the DataFrame
covariance_data = df[covariance_columns]

# Compute the covariance matrix
covariance_matrix = np.cov(covariance_data, rowvar=False)

covariance_matrix
# Compute the eigenvalues
eigenvalues, _ = np.linalg.eig(covariance_matrix)

# Sort eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]

# Plot the zoomed-in scree plot
plt.figure(figsize=(14, 8))  # Adjust the figsize parameter for a larger plot
plt.bar(range(1, len(sorted_eigenvalues) + 1), sorted_eigenvalues)
plt.xlabel('Principal Components')
plt.ylabel('Eigenvalues')
plt.title('Zoomed-In Scree Plot')

# Set x-axis limits to zoom in on the first 10 principal components, for example
plt.xlim(0, 10)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Assuming df is your DataFrame with numeric data


# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=np.number).columns
df_numeric = df[numeric_columns]

# Apply PCA
pca = PCA()
principal_components = pca.fit_transform(df_numeric)

# Extract the weights for the 1st and 2nd principal components
weights_pc1 = pca.components_[0]
weights_pc2 = pca.components_[1]

# Create a DataFrame to display the weights
weights_df = pd.DataFrame({'Feature': df_numeric.columns, 'PC1_Weights': weights_pc1, 'PC2_Weights': weights_pc2})

# Display the DataFrame with weights
weights_df
```

```python
import pandas as pd

# Assuming df is your DataFrame
weights_df['Feature'] = weights_df['Feature'].str.replace('_Log_Return', '')

# Display the modified DataFrame
weights_df
```

```python
data_company = pd.read_csv('SP500_ticker.csv', encoding='latin')
data_company 
```

```python
import pandas as pd

# Sample data for table 1
data1 = weights_df

# Sample data for table 2
data2 = data_company

# Create DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Merge the two DataFrames based on the 'ticker' column
result_df = pd.merge(df1, df2[['ticker', 'sector']], left_on='Feature', right_on='ticker', how='left')

# Drop the duplicate 'ticker' column
result_df = result_df.drop(columns='ticker')

# Display the resulting DataFrame

result_df
```

```python
import seaborn as sns

# Assuming 'sector' is a column in your DataFrame
# If it's not already, you can add it by using the following:
# results_df['sector'] = ...  # Replace with the actual column in your dataset

# Create a bar plot for the weights of the 1st principal component
plt.figure(figsize=(12, 8))
sns.barplot(x='sector', y='PC1_Weights', data=result_df, palette='viridis')
plt.title('Weights of the 1st Principal Component by Industry Sector')
plt.xlabel('Industry Sector')
plt.ylabel('PC1 Weight')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
